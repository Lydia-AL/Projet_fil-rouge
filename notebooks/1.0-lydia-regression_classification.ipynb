```python
data_3['Labels_Ewltp'] = data_3['Ewltp (g/km)'].apply(lambda x: 'A' if x <= 100 else ('B' if x <= 120 else ('C' if x <= 140 else ('D' if x <= 160 else ('E' if x <= 200 else ('F' if x <= 250 else ('G' if x > 250 else None)))))))
print(data_3)
```

```python
data_3['Labels_Ewltp'].unique()
```

```python
to_drop_regression = ['Labels_Ewltp', 'Ewltp (g/km)']
data_3_regression = data_3.drop(to_drop_regression, axis=1)
target_regression = data_3['Ewltp (g/km)']

to_drop_classification = ['Labels_Ewltp', 'Ewltp (g/km)']
data_3_classification = data_3.drop(to_drop_classification, axis=1)
target_classification = data_3['Labels_Ewltp']
```

```python
from sklearn.model_selection import train_test_split
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(data_3_regression, target_regression, test_size=0.20)
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(data_3_classification, target_classification, test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler
cols_num = ['Mt', 'W (mm)', 'At1 (mm)', 'ec (cm3)', 'ep (KW)', 'Erwltp (g/km)', 'Fuel consumption ']

sc = StandardScaler()
X_train_r[cols_num] = sc.fit_transform(X_train_r[cols_num])
X_test_r[cols_num] = sc.transform(X_test_r[cols_num])

X_train_c[cols_num] = sc.fit_transform(X_train_c[cols_num])
X_test_c[cols_num] = sc.transform(X_test_c[cols_num])
```

```python
display(X_train_r)
display(X_test_r)
display(X_train_c)
display(X_test_c)
display(y_train_r)
display(y_test_r)
display(y_train_c)
display(y_test_c)
```

```python
#Classification _ Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state=123)
dt_clf.fit(X_train_c, y_train_c)

y_pred_c = dt_clf.predict(X_test_c)
pd.crosstab(y_test_c, y_pred_c, rownames=['Classe rÃ©elle'], colnames=['Classe prÃ©dite'])
```

```python
#Calculs des scores
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

accuracy = accuracy_score(y_test_c, y_pred_c)
print(f"Accuracy: {accuracy:.4f}")
print("/n")
precision = precision_score(y_test_c, y_pred_c, average='weighted')
print(f"Precision: {precision:.4f}")
print("/n")
recall = recall_score(y_test_c, y_pred_c, average='weighted')
print(f"Recall: {recall:.4f}")
print("/n")
f1 = f1_score(y_test_c, y_pred_c, average='weighted')
print(f"F1 Score: {f1:.4f}")
print("/n")
conf_matrix = confusion_matrix(y_test_c, y_pred_c)
print("Confusion Matrix:")
print(conf_matrix)
print("/n")
class_report = classification_report(y_test_c, y_pred_c)
print("Classification Report:")
print(class_report)
```
